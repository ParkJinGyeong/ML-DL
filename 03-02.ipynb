```python
import pandas as pd
```

```python
df = pd.read_csv('data/Fish.csv')
```

```python
df.head()
```

```python
cond = df['Species'] == 'Perch'
```

```python
p_length = df.loc[cond, ['Length2']]
p_weight = df.loc[cond, ['Weight']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target = train_test_split(p_length[['Length2']], p_weight[['Weight']])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(train_input, train_target)
```

```python
lr.score(test_input, test_target)
```

```python
lr.predict([[100]])
```

```python
lr.coef_
```

```python
a = lr.coef_[0][0]
a
```

```python
b = lr.intercept_[0]
b
```

```python
lr.intercept_
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(train_input['Length2'], train_target['Weight'])
plt.plot([10, 50], [10*a + b, 50*a + b])
plt.show()
#학습을 위한 데이터셋 
```

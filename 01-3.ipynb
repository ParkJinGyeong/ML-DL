```python
#빙어 데이터를 가지고 오기 
smelt_df = df[df['Species'] == 'Smelt']
smelt_df.head()
```

```python
# to list로 형 변환 
smelt_df = df[df['Species'] == 'Smelt']
smelt_length = smelt_df['Length2'].tolist()
smelt_weight = smelt_df['Weight'].tolist()
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 사이킷런으로 조합을 만들어줌 
# k 최근접 
# 아까 도미와 빙어로 나눔 
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

```python
list(zip(length, weight)) 
#length와 weight를 하나로 묶어줌 
#튜플 형식으로 만들어줌 
#사이킷런은 리스트를 기대해줌
#그래서 리스트로 바꾼 상황 
```

```python
# 2차원으로 바꿔줘야 함 
fish_data = [[l, w]for l, w in zip(length, weight)]
fish_data
#도미와 빙어가 섞인 상태 
# for문 활용해서 [l, w] 형태로 만들어준 것 
# zip의 튜플을 list로 바꿔주는 것 
# 2차원 데이터를 안 받은 상황, 데이터 프레임 형식이면 굳이 필요 없는 2차원화의 과정이긴 함 
```

```python
target_data = [1] * len(bream_length) + [0] * len(smelt_length)
```

```python
from sklearn.neighbors import KNeighborsClassifier
#함수가 아니라 클래스라는 뜻임 
```

```python
kn = KNeighborsClassifier()
```

```python
#학습하려는 데이터, 그리고 정답지를 줘야함
#지도학습 
kn.fit(fish_data, target_data)
```

```python
kn.score(fish_data, target_data)
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(20, 400, c = 'black', s =100, marker = "*") #하나의 점만 찍고 싶을 때, 색 조절, 사이즈 가능  
# 마커 사용 가능 
plt.scatter(smelt_length, smelt_weight)
```

```python
kn.predict([[20,400]]) #대괄호 두 개 # 도미라는 것을 말해주는거임 
```

```python
kn.predict([[16,30]]) #빙어로 predict 
```

```python
#predict 넣은 값 중 가까운 값 분석해서 하는 것 # 문제점이 많은 모델이긴 함
#모든 생선을 다 참고하는 것 
```

```python
kn49 = KNeighborsClassifier(49) #49개 데이터를 전부다 비교해주세요, 
#근데 도미 데이터가 35라서 점을 어디를 찍든지 도미로 판단할 것임. 모든 데이터를 도미로 선택할 것임 
```

```python
kn49.fit(fish_data, target_data) #원래 있는 데이터와 정답지 데이터로 학습을 시키는 것 
```

```python
kn49.score(fish_data, target_data) #맞출확률이 70퍼라는 뜻 #하이퍼 파라미터라고 많이 부름 #35/49랑 같음 #모든 데이터를 이웃으로 참고해서 좋을게 없음
```

```python
kn = KNeighborsClassifier()
kn.fit(fish_data, target_data)

for n in range (5, 50):
    kn.n_neighbors = n

    score = kn.score(fish_data, target_data)

    print(n,score)
    if score < 1 : 
        break
```
